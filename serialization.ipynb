```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
import numpy as np
def process_data(X):
    X = np.delete(X, 1, 1)
    return X
```

```python
X = df.drop(labels='Outcome', axis=1).values
y = df[['Outcome']].values
```

```python
X, y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train = process_data(X_train)
```

```python
X_train.shape
```

# Train and Serialization

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
with open('artifacts/model.pkl', 'wb') as file:
    pickle.dump(model, file)
```

# Deserialization and Test

```python
with open('artifacts/model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

```python
loaded_model
```

```python
X_test = process_data(X_test)
```

```python
predictions = loaded_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
accuracy
```

# Encapsular procesado y modelo

```python
class MyModel:

    def __init__(self):
        self.model = LogisticRegression()
    
    @staticmethod
    def __process_data(X):
        X = np.delete(X, 1, 1)
        return X
    
    def train_model(self, X, y):
        X = self.__process_data(X)
        self.model.fit(X, y)

    def predict_data(self, X):
        X = self.__process_data(X)
        predictions = self.model.predict(X)
        return predictions
```

```python
model = MyModel()
```

```python
model
```

```python
model.train_model(X_train, y_train)
```

```python
model.model
```

```python
with open('artifacts/encapsulated_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
with open('artifacts/encapsulated_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

```python
predictions = loaded_model.predict_data(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
accuracy
```
